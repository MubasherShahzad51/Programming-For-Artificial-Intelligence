# **Working with Text Files**
---

- **Writing Text Files**

```python
# When File is not present it will create a new file
myFile1=open("file1.txt","w")
myFile1.write("Hello ")
myFile1.close()

# Writing Multiple Lines
myFile2=open("file2.txt","w")
myFile2.write("Hello \n")
myFile2.write("World")
myFile2.close()

# When File is present it will overwrite the file
myFile3=open("file3.txt","w")
myFile3.write("Mubashir Shahzad")
myFile3.close()

# Appending in File
myFile4=open("file3.txt","a")
myFile4.write(" loves girls & the girls are:\n")
myFile4.close()


# Writing Multiple Lines
girls=["1) Ayesha\n","2) Sana\n","3) Saba\n","4) Saima\n"]
myFile5=open("file3.txt","a")
myFile5.writelines(girls)
myFile5.writelines('''5) Huma
6) Hina
7) Hira
8) Hamna
9) Hania
10) Aiman''')
myFile5.close()



```

- **Reading Text Files**

```python
# Reading Complete File
myFile1=open("file3.txt","r")
print(myFile1.read())
print('-'*50)
myFile1.close()

# Reading Specific Number of Characters
myFile1=open("file3.txt","r")
print(myFile1.read(70))
print('-'*50)
myFile1.close()

# Reading Specific Number of Lines
myFile1=open("file3.txt","r")
print(myFile1.readline(),end='')
print(myFile1.readline(),end='')
print(myFile1.readline(),end='')
print('-'*50)
myFile1.close()

# Reading Complete File
myFile1=open("file3.txt","r")

while True:
    line=myFile1.readline()
    if line=='':
        break
    print(line,end='')
print('\n')
print('-'*50)
myFile1.close()
```

- **Using Context Manager (with)**

```python
# Writing Multiple Lines
list=['\n11) Lubna\n','12) Laiba\n','13) Laraib\n','14) Laila\n','15) Zainab\n','16) Zara\n','17) Zoya\n','18) Zunaira\n','19) Fatima\n','20) Amber\n']
with open("file3.txt","a") as myFile1:
    myFile1.writelines(list)

# Reading Complete File

with open("file3.txt","r") as myFile1:
    print(myFile1.read())
    print('-'*50)
    

# Seek Function

with open("file3.txt","r") as myFile1:
    myFile1.seek(45)   # 45 is the index number of the character from where we want to read the file
    print(myFile1.read(10))
    print(myFile1.tell())     # tell function tells the index number of the character from where we are reading the file
```

- **Working with XML files**

```python
import xml.etree.ElementTree as ET
import pandas as pd
tree=ET.parse("family.xml")
root=tree.getroot()

childrenList={}


for parent in root:
    childrenList[parent.tag]=[]
    for child in parent:
       childrenList[parent.tag].append(child.text)

# childrenList

df=pd.DataFrame(childrenList)
df.to_csv("family.csv",index=False)
df
```
