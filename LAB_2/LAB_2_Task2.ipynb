```python
print("Welcome To Chatbot")
responses = {
    "bmw": "This is BMW Car",
    "manufacturing": "It is manufacturing in Germany",
    "help": "How can I assist you?",
}
def response(user_input):
    if user_input.lower() in responses:
        return responses[user_input.lower()]
    else:
        return "I'm sorry, I don't understand that. Please put correct information as given above."

def chatbot():
    print("I'm here to help you. Or you may type 'bye', 'exit', or 'quit' to exit.")
    
    user_input = input("User: ").lower()

    if user_input in ('bye', 'exit', 'quit'):
        print("Goodbye!")
        return

    response = responses.get(user_input, "I'm sorry, I don't understand that.")
    print(response)
    
    chatbot() # recursively call
chatbot()
```

# Task 2
* Create a text based adventure game with a unique story line and
different unique paths where user roams a forest and can add,
remove or access items from his inventory

```python
import random

inventory = []
current_location = "forest_entrance"
forest = {
    "forest_entrance": {
        "desc": "You stand at the forest entrance. Three paths lead deeper into the woods.",
        "paths": {"left": "A clear stream flows to your left.",
                  "center": "The central path disappears into thick woods.",
                  "right": "A wooden hut is nestled to your right."},
        "items": [],
    },
    "left": {
        "desc": "You venture left and find a shiny object by the stream.",
        "paths": {"forest_entrance": "Return to the entrance.", "center": "Continue deeper into the forest."},
        "items": ["shiny object"],
    },
    "center": {
        "desc": "You follow the central path, surrounded by tall trees.",
        "paths": {"forest_entrance": "Return to the entrance.", "left": "Return to the stream.", "right": "Enter the hut."},
        "items": ["old map"],
    },
    "right": {
        "desc": "You take the right path and enter the wooden hut.",
        "paths": {"forest_entrance": "Return to the entrance.", "center": "Return to the central path."},
        "items": ["small key"],
    },
}

def add_item(item):
    inventory.append(item)

def remove_item(item):
    if item in inventory:
        inventory.remove(item)
        return True
    return False

def access_inventory():
    return inventory

print("Welcome to the Forest Adventure Game!")
print("You find yourself at the forest entrance.")
print("Explore, collect items, and enjoy your adventure.")
print("Type 'help' if you require.")

while True:
    print(forest[current_location]["desc"])
    action = input("What will you do? ").lower()

    if action == "help":
        print("Commands: 'go [direction]', 'inventory', 'take [item]', 'drop [item]', 'quit'")
    elif action.startswith("go "):
        direction = action[3:]
        if direction in forest[current_location]["paths"]:
            current_location = direction
        else:
            print("You can't go that way.")
    elif action == "inventory":
        print("Inventory:", ", ".join(access_inventory()))
    elif action.startswith("take "):
        item = action[5:]
        if item in forest[current_location]["items"]:
            add_item(item)
            forest[current_location]["items"].remove(item)
            print(f"You take the {item}.")
        else:
            print("There's no such item here.")
    elif action.startswith("drop "):
        item = action[5:]
        if remove_item(item):
            forest[current_location]["items"].append(item)
            print(f"You drop the {item}.")
        else:
            print("You don't have that item in your inventory.")
    elif action == "quit":
        print("Thanks for playing!")
        break
    else:
        print("Invalid command.")
```
