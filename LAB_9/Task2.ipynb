# **Making DataFrame For Columns Date Revenue & Expenses**

```python
import pandas as pd
import numpy as np

dates = np.array(['2022-01-01', '2022-02-01', '2022-03-01', '2022-04-01', '2022-05-01', '2022-06-01', '2022-07-01', '2022-08-01', '2022-09-01', '2022-10-01'])
revenues = np.array([50000, 55000, 60000, 65000, 70000, 75000, 80000, 85000, 90000, 95000])
expenses = np.array([30000, 32000, 35000, 38000, 40000, 42000, 45000, 48000, 50000, 52000])


df = pd.DataFrame({'Date': dates, 'Revenue': revenues, 'Expenses': expenses})
df
```

# Writing Data to CSV File

```python
df.to_csv('Financial_Data.csv', index=False)
```

# Importing Data From CSV File and Performing Operations Using Numpy Methods

```python
df = pd.read_csv("./Financial_Data.csv")

profits = revenues - expenses
average_profit = np.mean(profits)
```

# **Making New Column For Monthly Profit**

```python
df["Monthly Profit"] = pd.DataFrame(profits)
df
```

# **Making New Column For Average Profit**

```python
df["Average Profit"] = average_profit
df
```
